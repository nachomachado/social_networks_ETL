```python
from typing import Union
import pandas as pd
from fastapi import FastAPI
from pytrends.request import TrendReq
import collections
from datetime import date
```

```python
app = FastAPI()
pytrend = TrendReq()
df= pytrend.trending_searches()
```

```python
def letter_count(df,API_name):
    today=date.today()
    dict = {}
    list_API = df[0].tolist()
    str_API=''.join(list_API).lower()     #eliminar espacios
    str_API=str_API.replace(" ","")
    for i in str_API:
        dict[i] = str_API.count(i)
    df_final=pd.DataFrame.from_dict(dict, orient='index')
    df_final=df_final.sort_index().rename(columns={0:'char_count'})
    df_final.to_csv(f'../stagging/API_{API_name}/{today}_char_count.csv')
    return df_final
```

```python
letter_count(df,'google')
```
